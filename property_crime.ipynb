```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
d = pd.read_csv('data2020.csv')
```

```python
d.info()
```

```python
d.describe()
```

```python
d.head()
```

```python
plt.pie(d['Total'], labels = d['Area'], autopct = "%.2f%%", rotatelabels = True )
```

```python
cmap = plt.get_cmap('tab20b')
mycolors = cmap(np.arange(22))
```

```python
plt.pie(d['Total'], labels = d['Area'], autopct = "%.2f%%", wedgeprops = dict(width = 0.3),
        colors = mycolors, rotatelabels = True)
plt.title("Total property crime in Delhi area-wise in 2020")

sns.set(rc = {'figure.figsize':(10,10)})
```

```python
plt.bar(d['Area'], d['Total'], color = "Red" )
plt.xticks(rotation = 90)
plt.title("Total property crime in Delhi area-wise in 2020")
```

```python
d.plot.area()
plt.title('Crime in Delhi category wise in 2020')
```

```python
plt.stackplot(d['Area'], d['Robbery']/d.sum(axis = 1), d['Theft']/d.sum(axis = 1),
              d['Burglary']/d.sum(axis = 1), d['Dacoity']/d.sum(axis = 1),
              colors = ['yellow','green','brown','red'],
              labels = ['Robbery','Theft','Burglary','Dacoity'])
plt.legend();
plt.xticks(rotation = 90)
```

```python
sns.barplot(y = 'Area', x = 'Total', data = d)
plt.xticks(rotation = 90)
plt.title('Crime in Delhi category wise in 2020')
```

```python
sns.heatmap(d)
```
